```python
def schema1(a, U0):
    n = len(U0)
    U1 = U0 - a * (U0 - np.concatenate([[U0[-1]],U0[:n-1]])) 
    return U1

"""def schema2(a, U0):
    n = len(U0)
    U1 = U0 - a/2 * (np.concatenate([U0[1:], [U0[0]]])
                     - np.concatenate([[U0[-1]],U0[:n-1]])) 
    + a**2*(np.concatenate([U0[1:],[U0[0]]])
            - 2*U0 + np.concatenate([[U0[-1]],U0[:n-1]]))/2
    return U1"""

def schema2(a, U0):
    n = len(U0)
    U1 = np.zeros(n)
    U1[0] = U0[0] - a * (U0[1] - U0[-1])/2 + a**2*(U0[1] - 2*U0[0] + U0[-1])/2
    U1[1:n-1] = U0[1:n-1] - a * (U0[2:] - U0[:n-2])/2 + a**2*(U0[2:] - 2 * U0[1:n-1] + U0[:-2])/2
    U1[n-1] = U0[n-1] - a * (U0[0] - U0[n-2])/2 + a**2*(U0[0] - 2*U0[n-1] + U0[n-2])/2
    return U1

def schema3(a, U0):
    n = len(U0)
    U1 = U0 - a * (U0 - np.concatenate([[U0[-1]],U0[:n-1]])) 
    - a*(1-a)/4 * (np.concatenate([U0[1:],[U0[0]]]) 
                   - U0 - np.concatenate([[U0[-1]],U0[:n-1]]) 
                   + np.concatenate([[U0[-2],U0[-1]],U0[:n-2]]))
    return U1
```

```python
def condition1 (I):
    t = np.linspace(-1,1,2*I + 1)[:-1]
    X = 256 * (t - 0.25)**2 * (t + 0.25)**2 * (t >= -0.25) * (t <= 0.25)
    return X
```

```python
def calcul(sch, T, I, a):
    u = condition1(I)
    if sch == 1:
        for i in range(T):
            u = schema1(a, u)
    if sch == 2: 
        for i in range(T):
            u = schema2(a, u)
    if sch == 3: 
        for i in range(T):
            u = schema3(a, u)
    return u
```
