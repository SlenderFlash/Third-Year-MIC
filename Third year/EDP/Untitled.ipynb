1. Equation de Poisson:
<br> Nous avons $-u''(x) = f(x) = \pi^2sin(\pi x)$ donc $u(x) = sin(\pi x) + ax + b$ mais $u(0) = u(1) = 0$ donc $u(x) = sin(\pi x)$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def f(x):
    return np.pi**2*np.sin(np.pi*x)
def u(x):
    return np.sin(np.pi*x)
```

```python
for n in range(3,8):
    abscisse = np.linspace(0,1,2**n+1)
    plt.figure()
    uo = u(abscisse)
    h = 1/2**n
    F = f(abscisse)
    A = (np.diag(np.full(2**n+1,2))-np.diag(np.ones(2**n),1)-np.diag(np.ones(2**n),-1))/h**2
    U = np.linalg.solve(A,F)
    plt.plot(abscisse,U, label = 'approx')
    plt.plot(abscisse,uo, label = 'exact')
    plt.legend()
    plt.show()
    print(max(abs(uo-U)))
    print(np.linalg.norm(uo-U, ord = 2))
    
```

```python
def f(x):
    return 1 if (x > 0 and x < 1/2) else 0
def u(x):
    if (x < 1/2 and x >= 0):
        result = 3*x/8 - x**2/2
    else:
        result = 1/8 - x/8
    return result
```

```python
for n in range(5,10):
    abscisse = np.linspace(0,1,2**n+1)
    plt.figure()
    uo = [u(x) for x in abscisse]
    h = 1/2**n
    F = [f(x) for x in abscisse]
    A = (np.diag(np.full(2**n+1,2))-np.diag(np.ones(2**n),1)-np.diag(np.ones(2**n),-1))/h**2
    U = np.linalg.solve(A,F)
    plt.plot(abscisse,U, label = 'approx')
    plt.plot(abscisse,uo, label = 'exact')
    plt.legend()
    plt.show()
    print(max(abs(uo-U)))
    print(np.linalg.norm(uo-U, ord = 2))
```

```python
def f(x):
    return 1 if (x > 0 and x < 1/2) else 0
def u(x):
    if (x < 1/2 and x >= 0):
        result = 11*x/8 - x**2/2
    else:
        result = 1/8 + 7*x/8
    return result
```

```python
for n in range(5,10):
    abscisse = np.linspace(0,1,2**n+1)
    plt.figure()
    uo = [u(x) for x in abscisse]
    h = 1/2**n
    F = [f(x) for x in abscisse]
    A = (np.diag(np.full(2**n+1,2))-np.diag(np.ones(2**n),1)-np.diag(np.ones(2**n),-1))/h**2
    U = np.linalg.solve(A,F) + abscisse
    plt.plot(abscisse,U, label = 'approx')
    plt.plot(abscisse,uo, label = 'exact')
    plt.legend()
    plt.show()
    print(max(abs(uo-U)))
    print(np.linalg.norm(uo-U, ord = 2))
```
