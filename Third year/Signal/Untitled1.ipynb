```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wv
```

```python
%matplotlib inline
```

```python
def CreateTrame(x,w,R = 0.5):
    n = len(x)
    nw = len(w)
    step = int(nw*(1-R))
    nb = int((n-nw)/step)+1
    B = np.zeros((nb,nw))
    for ind in range(nb):
        B[ind,:] = x[ind*step:nw+ind*step]*w
    return B
```

```python
Freq, Signal = wv.read('bubbles.wav')
Hann240 = np.hanning(240)
T = 3
f_echan = 8000
f = 20
n_echan = f_echan*T+1
temp = np.linspace(0,3,n_echan)
sinus = np.sin(2*np.pi*f*temp)
Decom = CreateTrame(sinus, Hann240)
print(Freq)
for i in range(3):
    plt.figure()
    plt.plot(Decom[i])
    plt.show()
```

```python
 def AddTrame(B,w,R = 0.5):
    (nb, nw) = np.shape(B)
    step = int(nw*(1-R))
    n = nb*step + nw
    x = np.zeros(n)
    n_recouvre = int(1/(1-R))+1
    for ind in range(n-nw):
        up_b = int(ind/step)
        be_b = int(max(np.ceil((ind+1-nw)/step),0))
        for t in range(be_b,up_b+1):
            x[ind] += B[t,ind-t*step]
    return x
```

```python
Recon = AddTrame(Decom,Hann240)
plt.plot(Recon)
```
